```python
def OK_GetRelativesV2(uid):
    method = 'friends.getRelativesV2'
    uid = uid
    reltypes = 'ALL'
    method_kwargs = {'uid': uid, 'reltypes':reltypes}
    url = MakeRequestURL(method, **method_kwargs)
    req = get(url)
    answer = json.loads(req.text)
    return answer


def GetFriendsListID(uid):
    method = 'friends.get'
    fid = uid
    method_kwargs = {'fid': fid}
    url = MakeRequestURL(method, **method_kwargs)
    req = get(url)
    answer = json.loads(req.text)
    return answer

def URL2UID(url):
    method = 'url.getInfo'
    method_kwargs = {'url': url}
    url = MakeRequestURL(method, **method_kwargs)
    req = get(url)
    answer = json.loads(req.text)
    if answer['type'] == 'PROFILE':
        return str(answer['objectId'])
    else:
        print(answer['type'])
        return str(answer['objectId'])
    
def GetInfo(uids):
    method = 'users.getInfo'
    fields = ['AGE', 'BIRTHDAY', 'EMAIL', 'FIRST_NAME', 'GENDER', 'HAS_EMAIL',
             'LAST_NAME', 'LOCALE', 'LOCATION', 'NAME', 'ODKL_EMAIL', 
             'ODKL_LOGIN', 'RELATIONS', 'RELATIONSHIP', 'UID',
             'URL_PROFILE']
#     uids = '119904479500'
    method_kwargs = {'uids': uids, 'fields': fields}
    url = MakeRequestURL(method, **method_kwargs)
    req = get(url)
    answer = json.loads(req.text)
    try:
        yyyy, mm, dd = answer[0]['birthday'].split('-')
        answer[0]['birthday'] = dd + '.' + mm + '.' + yyyy
    except:
        try:
            mm, dd = answer[0]['birthday'].split('-')
            answer[0]['birthday'] = dd + '.' + mm
        except:
            pass
#             print('ERROR in Ok birthday')
    try:
        return answer[0]
    except:
        return answer

def GetBirthday(uids):
    method = 'users.getInfo'
    fields = ['AGE', 'BIRTHDAY']
#     uids = '119904479500'
    method_kwargs = {'uids': uids, 'fields': fields}
    url = MakeRequestURL(method, **method_kwargs)
    req = get(url)
    answer = json.loads(req.text)
    try:
        yyyy, mm, dd = answer[0]['birthday'].split('-')
        answer[0]['birthday'] = dd + '.' + mm + '.' + yyyy
    except:
        try:
            mm, dd = answer[0]['birthday'].split('-')
            answer[0]['birthday'] = dd + '.' + mm
        except:
            pass
#             print('ERROR in Ok birthday')
    try:
        return answer[0]['birthday']
    except:
        return None
    
def MakeRequestURL(method, ok=ok, **kwargs):
    request_ = DotMap()
    
    parameters = ['application_key={}'.format(ok.OAuthOK['APP_PUBLIC']),
                  'format={}'.format(ok.format),
                  'method={}'.format(method),
#                   '{}'.format(ok.access_info['session_secret_key']),
                 ]

    for key in kwargs.keys():
        if type(kwargs[key]) == list:
            s = key
            for el in kwargs[key]:
                if len(s) != len(key):
                    s += ', '
                else:
                    s += '='
                s += el
        else:
            s = key + '=' + kwargs[key]
        parameters.append(s)
        
    parameters.sort()
#     print(parameters)
    parameters_str = ''
    for el in parameters:
        parameters_str += el
    parameters_str += ok.access_info['session_secret_key']
    
    request_.sig = GetMD5(parameters_str)
    
#     print(parameters)
#     print()
#     print(parameters_str)
#     print()
#     print('sig: ', request_.sig)
    
    kwargs_str = ''
    for key in kwargs.keys():
        if key == 'url':
            url = urllib.parse.quote(kwargs[key], safe='')
            kwargs_str += '&' + key + '=' + url
        if type(kwargs[key]) == list:
            s = ''
            for el in kwargs[key]:
                if len(s) != 0:
                    s += ', '
                s += el
            kwargs_str += '&' + key + '=' + urllib.parse.quote(s, safe='')
        else:
            kwargs_str += '&' + key + '=' + kwargs[key]
    
    request_.request = '''https://api.ok.ru/fb.do?
application_key={}
&format={}
&method={}
{}
&sig={}
&access_token={}'''.format(
        ok.OAuthOK['APP_PUBLIC'],
        ok.format,
        method,
        kwargs_str,
        request_.sig,
        ok.access_info['access_token'] )
    
    request_.request = request_.request.replace('\n', '')
    
#     print(request_.request)
    return request_.request
```
