```python
class Person(object):
    def __init__(self, first_name, last_name, birthday=None,
                 city=None, country=None, vk_id=None, ok_id=None, vk_link=None):
        self.first_name = first_name
        self.last_name = last_name
        self.birthday = birthday
        self.age = self.calculate_age()
        self.city = country
        self.city = city
        self.career = None
        self.career_time = None
        self.about = None
        self.activities = None
        self.home_phone = None
        self.mobile_phone = None
        self.schools = None
        self.faculty = None
        self.universities = None
        self.faculty_name = None
        self.university_name = None
        self.graduation = None
        self.interests = None
        
        self.vk_id = vk_id
        self.vk_link = vk_link
        self.vk_friends_list = [] #friend = (Person)
        self.vk_friends_list_2 = []
        self.vk_photo_link = None    
        
        self.ok_id = ok_id
        self.ok_link = None
        self.ok_friends_list = []
        self.ok_additional_info = []
        
        self.ok_BROTHERSISTER = None
        self.ok_CHILD = None
        self.ok_CHILDINLAW = None
        self.ok_CLASSMATE = None
        self.ok_CLOSEFRIEND = None
        self.ok_COLLEGUE = None
        self.ok_COMPANIONINARMS = None
        self.ok_CURSEMATE = None
        self.ok_GODCHILD = None
        self.ok_GODPARENT = None
        self.ok_GRANDCHILD = None
        self.ok_GRANDPARENT = None
        self.ok_LOVE = None
        self.ok_NEPHEW = None
        self.ok_PARENT = None
        self.ok_PARENTINLAW = None
        self.ok_PLAYINGTOGETHER = None
        self.ok_RELATIVE = None
        self.ok_SPOUSE = None
        self.ok_UNCLEAUNT = None
        
        self.friends_list = None
        
        self.vk_sole = True
    
    def calculate_age(self):
        if self.birthday!=None and len(self.birthday.split('.')) == 3:
            today = date.today()
            age = today.year - int(self.birth_year()) - ((today.month, today.day) < (int(self.birth_month()), int(self.birth_day())))
            return int(age)
        else:
            return None
        

    def print_info(self):
#         attributes = [attr for attr in dir(self) if not attr.startswith('__')]
        for key in self.__dict__.keys():
            if self.__dict__[key] != None:
                if type(self.__dict__[key]) == list:
                    if len(self.__dict__[key]) !=0:
                        print(key, ': ', self.__dict__[key])
                else:
                    print(key, ': ', self.__dict__[key])
                    
    def to_dict(self):
        person_dict = {}
        for key in self.__dict__.keys():
            if key not in ['vk_friends_list', 'ok_friends_list']:
                person_dict[key] =  self.__dict__[key]
        return person_dict
    
    def to_dict_without_None(self):
        person_dict = {}
        for key in self.__dict__.keys():
            if key not in ['vk_friends_list', 'ok_friends_list']:
                if self.__dict__[key] == None:
                    person_dict[key] =  ''
                else:
                    person_dict[key] =  self.__dict__[key]
        return person_dict

    def person(self):
        return(str(self))
    
    def __repr__(self,):
        if self.birthday == None and self.vk_link is not None:
            return '{} {} {}'.format(self.last_name, self.first_name, self.vk_link)
        return '{} {} {}'.format(self.last_name, self.first_name, self.birthday)

    def name(self):
        return '{} {}'.format(self.last_name, self.first_name)
    
    def birth_day(self):
        return self.birthday.split('.')[0]
    
    def birth_month(self):
        return self.birthday.split('.')[1]
    
    def birth_year(self):
        return self.birthday.split('.')[2]
    
    def vk_get_info(self):
        pass
```

```python
def MergeFriends(person):
    friends_list = []
    friends_list_names = []

    index_ = 0
    for vk_id in person.vk_friends_list:
        index_+=1
        clear_output(True)
        print('MergeFriends (Vk):\nPerson: {}. Complete {} of {} links'.format(person, index_, len(person.vk_friends_list)))
        sleep(1)
        user_info = vk.api.users.get(user_id=vk_id,
                                     fields=['last_name', 'first_name', 'bdate'])[0]
        try:
            if 'bdate' not in user_info.keys():
                user_info['bdate'] = None
                name = user_info['last_name'] + ' ' + user_info['first_name']
            else:           
                name = user_info['last_name'] + ' ' + user_info['first_name'] + ' ' + user_info['bdate']

            ind = 0
            while name in friends_list_names:
                name += str(ind)
                ind += 1
                user_info['first_name'] += ' ' + str(ind)

            friends_list_names.append(name)
    #         print(user_info['bdate'])

            friends_list.append(Person(first_name=user_info['first_name'],
                                       last_name=user_info['last_name'],
                                       birthday=user_info['bdate'],
                                       vk_id=vk_id))
        except:
            pass
#             print('ERROR: vk MergeFriends')

    print('Vk: Done')
    index_ = 0
    print('MergeFriends (Ok):\nPerson: {}. Complete {} of {} links'.format(person, index_, len(person.ok_friends_list)))
    sleep(2)
    intersection_number = 0
    for ok_id in person.ok_friends_list:
        index_+=1
        sleep(0.1)
        clear_output(True)
        print('MergeFriends (Ok):\nPerson: {}. Complete {} of {} ok_friends_list'.format(person, index_, len(person.ok_friends_list)))        
        user_info = GetInfo(ok_id)
        try:
            name = user_info['last_name'] + ' ' + user_info['first_name'] + ' ' + user_info['birthday']
#             print(name)
            if name in friends_list_names:
                index = friends_list_names.index(name)
                friends_list[index].ok_id = ok_id
                intersection_number+=1
#                 print(friends_list[index])
            else:
                friends_list_names.append(name)
                friends_list.append(Person(first_name=user_info['first_name'],
                                           last_name=user_info['last_name'],
                                           birthday=user_info['birthday'],
                                           ok_id=ok_id))
                
        except:
            pass
#             print('ERROR: ok MergeFriends')
    print('Ok: Done')
    print('Intersection_number (vk,ok): ', intersection_number)
    return friends_list


def GetRelativesV2(person):
    for el in OK_GetRelativesV2(person.ok_id)['relations']:
        friends_list_names = []
        persons_list = []
        
        try:
            if el['type_id'] == 'COLLEGUE':
                for per_id in el['uids']:
                    per_url = GetInfo(per_id)['url_profile']
                    per_info = GetAdditionalInfoFromWeb(per_url)
    #                 print(per_info)
                    try:
                        if len(person.career) == 0:
                            person.career += per_info['организация']
                        else:
                            person.career += ' ' + '\\' + ' ' + per_info['организация']
                            break
                    except:
                        pass
        except:
            pass
        
        try:
            for uid in el['uids']:
                try:
                    user_info = GetInfo(uid)
                    name = user_info['last_name'] + ' ' + user_info['first_name'] + ' ' + user_info['birthday']
                    friends_list_names.append(name)
                    persons_list.append(Person(first_name=user_info['first_name'],
                                               last_name=user_info['last_name'],
                                               birthday=user_info['birthday'],
                                               ok_id=uid
                                              ))
                except:
                    pass
        except:
                    pass
        person.__dict__['ok_'+el['type_id']] = persons_list
#         print(persons_list)
#         print('---')
```
