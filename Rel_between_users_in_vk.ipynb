# Columns Renaming

```python
a_1_step.columns = ['Person_from_list_A','Friends_of_person_from_list_A']
a_2_step.columns = ['Friends_of_person_from_list_A','Friends_of_friends_of_person_from_list_A']
b_1_step.columns = ['Person_from_list_B','Friends_of_person_from_list_B']
b_2_step.columns = ['Friends_of_person_from_list_B','Friends_of_friends_of_person_from_list_B']
```

```python
a_1_step['Person_from_list_A'] = a_1_step['Person_from_list_A'].astype(int)
a_1_step['Friends_of_person_from_list_A'] = a_1_step['Friends_of_person_from_list_A'].astype(int)
a_2_step['Friends_of_person_from_list_A'] = a_2_step['Friends_of_person_from_list_A'].astype(int)
a_2_step['Friends_of_friends_of_person_from_list_A'] = a_2_step['Friends_of_friends_of_person_from_list_A'].astype(int)
```

```python
a_1_2_step = a_1_step.merge(a_2_step,how='inner')
a_1_2_step.head()
```

```python
b_1_step['Person_from_list_B'] = b_1_step['Person_from_list_B'].astype(int)
b_1_step['Friends_of_person_from_list_B'] = b_1_step['Friends_of_person_from_list_B'].astype(int)
b_2_step['Friends_of_person_from_list_B'] = b_2_step['Friends_of_person_from_list_B'].astype(int)
b_2_step['Friends_of_friends_of_person_from_list_B'] = b_2_step['Friends_of_friends_of_person_from_list_B'].astype(int)
```

```python
b_1_2_step = b_1_step.merge(b_2_step,how='inner')
b_1_2_step.head()
```

# Level 1

```python
rel_1_level = a_1_2_step.merge(b_1_2_step, 
                               left_on='Friends_of_person_from_list_A',
                               right_on='Person_from_list_B')
rel_1_level = rel_1_level[['Person_from_list_A',
             'Friends_of_person_from_list_A',
            'Person_from_list_B']].drop_duplicates()

rel_1_level['Person_from_list_A'] = 'vk.com/id' + rel_1_level['Person_from_list_A'].astype(str)
rel_1_level['Friends_of_person_from_list_A'] = 'vk.com/id' + rel_1_level['Friends_of_person_from_list_A'].astype(str)
rel_1_level['Person_from_list_B'] = 'vk.com/id' + rel_1_level['Person_from_list_B'].astype(str)
rel_1_level
```
